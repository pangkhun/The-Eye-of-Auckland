# Code to Clean the Raw Twitter Data

## Import Packages and Files

```python
#import necessary packages
import pandas as pd
import re
from datetime import datetime
```

```python
#import raw twitter data
file_location = 'C:/Users/pkhu624/OneDrive - The University of Auckland/399/KeywordFromLocation/solution.csv'
Location_tweets = pd.read_csv('solution.csv')
display(Location_tweets)

#import a list of locations and their latitude/longitude as a reference dataframe
map_data=pd.read_csv('position.csv')
display(map_data)

#Keyword Extraction file
Keyword_Extraction_Location = 'C:/Users/pkhu624/OneDrive - The University of Auckland/399/Insta+FB/Keywords Extraction Analysis.csv'
extraction_df = pd.read_csv(Keyword_Extraction_Location)
```

## Preprocessing

```python
#keep only unique tweets in dataset
unique_count = Location_tweets['rawContent'].nunique()
Location_tweets = Location_tweets.drop_duplicates(subset='rawContent')

#Use to drop columns from original dataset. 
cols_to_drop = ['Unnamed: 0', 'url', 'id', 'user', 'conversationId', 'lang', 'source', 'sourceUrl', 'sourceLabel', 'links', 'media', 'retweetedTweet', 'quotedTweet', 'inReplyToTweetId', 'inReplyToUser', 'mentionedUsers', 'card', 'vibe', 'bookmarkCount', 'place', 'renderedContent', 'cashtags']
Location_tweets.drop(cols_to_drop, axis=1, inplace=True)

#Get tweets that occured only after 2022
Location_tweets['date'] = pd.to_datetime(Location_tweets['date'])
Location_tweets = Location_tweets[Location_tweets['date'].dt.year > 2021]

#change strings to lower 
Location_tweets['rawContent'] = Location_tweets['rawContent'].str.lower()
Location_tweets['hashtags'] = Location_tweets['hashtags'].str.lower()
```

## Specify Functions used to process Dataframe

```python
#Function to only keep rows of dataframe that contains a location
def recognise_text(targetwords, dataframe, columns):
    #create a new dataframe instane
    filtered_dataframe = pd.DataFrame()
    
    for column in columns:
        # Use boolean indexing to filter rows containing any of the locations
        mask = dataframe[column].str.contains('|'.join(targetwords), case=False, na=False)
        
        #append the row that contains location to the new dataframe instance
        filtered_dataframe = filtered_dataframe.append(dataframe[mask])
    
    return filtered_dataframe
```

## Use Functions to Filter Incident Type

```python
#arguments to be passed into functions
columns_to_search = ['rawContent', 'hashtags']

#incident types
fire = (extraction_df.iloc[:,0].dropna()).tolist()
flood = (extraction_df.iloc[:,1].dropna()).tolist()
storm = (extraction_df.iloc[:,2].dropna()).tolist()
landslide = (extraction_df.iloc[:,3].dropna()).tolist()
harass = (extraction_df.iloc[:,4].dropna()).tolist()
battery = (extraction_df.iloc[:,5].dropna()).tolist()
theft = (extraction_df.iloc[:,6].dropna()).tolist()
burglary = (extraction_df.iloc[:,7].dropna()).tolist()
robbery = (extraction_df.iloc[:,8].dropna()).tolist()
sexual_assault = (extraction_df.iloc[:,9].dropna()).tolist()
```

```python
# find the rows that contain a recognised incident
filtered_fire = recognise_text(fire, Location_tweets, columns_to_search)
filtered_flood = recognise_text(flood, Location_tweets, columns_to_search)
filtered_storm = recognise_text(storm, Location_tweets, columns_to_search)
filtered_landslide = recognise_text(landslide, Location_tweets, columns_to_search)
filtered_harass = recognise_text(harass, Location_tweets, columns_to_search)
filtered_battery = recognise_text(battery, Location_tweets, columns_to_search)
filtered_theft = recognise_text(theft, Location_tweets, columns_to_search)
filtered_burglary =recognise_text(burglary, Location_tweets, columns_to_search)
filtered_robbery =recognise_text(robbery, Location_tweets, columns_to_search)
filtered_sexual_assault = recognise_text(sexual_assault, Location_tweets, columns_to_search)

#Add incident type to the rows 
filtered_fire['Incident'] = 'Fire'
filtered_flood['Incident'] = 'Flood'
filtered_storm['Incident'] = 'Storm'
filtered_landslide['Incident'] = 'Landslide'
filtered_harass['Incident'] = 'Harassment'
filtered_battery['Incident'] = 'Battery'
filtered_theft['Incident'] = 'Theft'
filtered_burglary['Incident'] = 'Burglary'
filtered_robbery['Incident'] = 'Robbery'
filtered_sexual_assault['Incident'] = 'Sexual Assault'


#concatonate rows into one combined dataframe
filtered_df = pd.concat([filtered_fire, filtered_landslide, filtered_flood, filtered_storm, filtered_harass, filtered_battery, filtered_sexual_assault, filtered_theft, filtered_burglary, filtered_robbery])
```

## Mapping location to Area Unit

```python
#Change columns to name to be in the same format 
new_column_names = {'date': 'Date', 'rawContent': 'Content','replyCount':'Comments','retweetCount':'Forward','likeCount':'Like'}
final_df = filtered_df.rename(columns=new_column_names)

pattern = r'\((.*?)\)'  # Regular expression pattern to match content inside parentheses
column_to_extract = 'Position'  # Column name to extract data from

# Apply regular expression matching and extract data inside parentheses
map_data[column_to_extract] = map_data[column_to_extract].apply(lambda x: re.findall(pattern, x)).apply(' '.join)
map_data['Area Unit'] = map_data['Area Unit'].apply(lambda x: x[:-1])
map_data

final_df.Location = final_df.Location.apply(lambda x: x.split(',')[:-1]).apply(','.join)
final_df

final_df = final_df.merge(map_data, left_on='Location', right_on='Position', how='left')
final_df

final_df[['Longitude', 'Latitude']] = final_df['Location'].str.split(',', expand=True)
```

## Reformatting Dataframe Structure

```python
#Change date to be in the same format as all other clean datasets
final_df['Date'] = pd.to_datetime(final_df['Date'])
final_df['date'] = final_df['Date'].dt.date
final_df['Hour'] = final_df['Date'].dt.strftime('%I:%M %p')
final_df['Occurrence Day Of Week'] = final_df['Date'].dt.strftime('%a')
final_df

#Reorder the columns to match other clean datasets
new_order = ['Hour', 'Occurrence Day Of Week', 'Incident', 'Area Unit', 'date', 'Date', 'Content', 'Comments', 'Forward', 'Like', 'Location',
         'Position', 'Longitude', 'Latitude']

# Reorder columns using reindex function
final_df = final_df.reindex(columns=new_order)

# Drop columns used to extract longitude/latitude and Area Unit
final_df.drop(['Location', 'Position', 'Date', 'Content'], axis=1, inplace=True)
final_df

# Rename columns to match names of other datasets
new_column_names = {'date': 'Date', 'Longitude': 'Latitude', 'Latitude': 'Longitude'}
final_df = final_df.rename(columns=new_column_names)
final_df

final_df['Number of Victimisations'] = 1

new_order = ['Hour', 'Occurrence Day Of Week', 'Incident', 'Area Unit', 'Date',
       'Comments', 'Forward', 'Like', 'Number of Victimisations', 'Latitude', 'Longitude']


# Reorder columns using reindex function
final_df = final_df.reindex(columns=new_order)

#drop unnecessary columns
cols_to_drop = ['Comments', 'Forward', 'Like'] 
final_df.drop(cols_to_drop, axis=1, inplace=True, )

```

## Saving Clean Data

```python
# Save the Dataframe to a CSV file
final_df.to_csv('Clean Twitter Data.csv', index=False)
display(final_df)
```
